# 2D Rotations

```julia
function rot2d(v::Matrix{T}, θ::Real) where T <: Real
    R= [
        cos(θ)  -sin(θ);
        sin(θ)  cos(θ)
    ]
    vr = R * v
   return vr 
end
```

```julia
v₀ = [0.0 30.0;
      0.0  0.0]
plt = plot(
    v₀[1,:], v₀[2,:],
    aspect_ratio = :equal,
    label = "v"
)

# plot a circle
angles = range(0, stop=2*π, length=1000)
r = sqrt(sum(v₀[:, 2].^2))
x = r * cos.(angles)
y = r * sin.(angles);
plot!(x, y, label="");
```

```julia
@manipulate for angle = 0:0.01:2π
    vr = rot2d(v₀, angle)
    
    plt.series_list[1].plotattributes[:x] = vr[1, :]
    plt.series_list[1].plotattributes[:y] = vr[2, :]
    
    plot!(plt)
end
```

```julia
# relative rotations
v = v₀
θ = 0.0
@manipulate for angle = 0:0.01:2π
    global v, θ
    vr = rot2d(v, angle - θ)
    v = vr
    θ = angle
    
    plt.series_list[1].plotattributes[:x] = vr[1, :]
    plt.series_list[1].plotattributes[:y] = vr[2, :]
    
    plot!(plt)
        
end
```

# Euler Angles
## Tait-Bryan Angles

```julia
function R_TaitBryan(;φ::Real, θ::Real, ψ::Real) 
    Rφ = [
        1 0 0;
        0  cos(φ) sin(φ);
        0 -sin(φ) cos(φ)
    ]
    Rθ = [
        cos(θ) 0 -sin(θ);
        0      1   0;
        sin(θ) 0  cos(θ)
    ]
    Rψ = [
        cos(ψ)  sin(ψ) 0; 
        -sin(ψ) cos(ψ) 0; 
        0 0 1
    ]

    R =   Rψ * Rθ * Rφ
    return R
end
```

```julia
function rot3d_TaitBryan(v::AbstractArray{T}; φ::Real, θ::Real, ψ::Real) where T <: Real    
    R = R_TaitBryan(φ=φ, θ=θ, ψ=ψ) 
    return  R * v
end
```

```julia
function rot3d(v::AbstractArray{T}, R::Matrix{Float64}) where T <: Real  
    size(R) == (3,3) || throw(ArgumentError("size(R)=$(size(R)), must be a (3, 3) matrix"))
    det(R) ≈ 1.0 || throw(ArgumentError("det(R)=$(det(R)), must be 1.0"))

    return  R * v
end
```

# Quaternions

```julia
import Base: convert, show
import Base: +, -, *, /, ^
import Base: conj, abs, abs2, inv
```

```julia
struct Quaternion{T<:Real} <: Number
    s::T
    v1::T
    v2::T
    v3::T
end

(::Type{Quaternion{T}})(x::Real) where {T} = Quaternion{T}(x, 0, 0, 0)
(::Type{Quaternion{T}})(q::Quaternion{T}) where {T<:Real} = q

Quaternion(x::Real) = Quaternion(x, zero(x), zero(x), zero(x))
Quaternion(s::Real, v1::Real, v2::Real, v3::Real) =
    Quaternion(promote(s, v1, v2, v3)...)
Quaternion(s::Real, a::Vector) = Quaternion(s, a[1], a[2], a[3])
Quaternion(a::Vector) = Quaternion(0, a[1], a[2], a[3])

convert(::Type{Quaternion{T}}, q::Quaternion) where {T} =
    Quaternion(convert(T, q.s), convert(T, q.v1), convert(T, q.v2), convert(T, q.v3) )
convert(::Type{Quaternion{T}}, q::Quaternion{T}) where {T <: Real} = q

function show(io::IO, q::Quaternion)
    pm(x) = x < 0 ? " - $(-x)" : " + $x"
    print(io, q.s, pm(q.v1), "im", pm(q.v2), "jm", pm(q.v3), "km")
end

# mathematical operations

conj(q::Quaternion) = Quaternion(q.s, -q.v1, -q.v2, -q.v3)
abs(q::Quaternion) = sqrt(q.s * q.s + q.v1 * q.v1 + q.v2 * q.v2 + q.v3 * q.v3)
abs2(q::Quaternion) =     q.s * q.s + q.v1 * q.v1 + q.v2 * q.v2 + q.v3 * q.v3

(+)(q::Quaternion, r::Quaternion) =
    Quaternion(q.s + r.s, q.v1 + r.v1, q.v2 + r.v2, q.v3 + r.v3)
(-)(q::Quaternion) = Quaternion(-q.s, -q.v1, -q.v2, -q.v3)
(-)(q::Quaternion, r::Quaternion) =
    Quaternion(q.s - r.s, q.v1 - r.v1, q.v2 - r.v2, q.v3 - r.v3)
(*)(q::Quaternion, r::Quaternion) = Quaternion(
        q.s * r.s - q.v1 * r.v1 - q.v2 * r.v2 - q.v3 * r.v3,
        q.s * r.v1 + q.v1 * r.s + q.v2 * r.v3 - q.v3 * r.v2,
        q.s * r.v2 - q.v1 * r.v3 + q.v2 * r.s + q.v3 * r.v1,
        q.s * r.v3 + q.v1 * r.v2 - q.v2 * r.v1 + q.v3 * r.s,
        )
(/)(q::Quaternion, x::Real) = Quaternion(q.s / x, q.v1 / x, q.v2 / x, q.v3 / x)
inv(q::Quaternion) = conj(q) / abs2(q)
;
```

```julia
function rot3d_quat(v::AbstractArray, n::AbstractVector, θ::Real)
    if length(n) != 3
        throw(ArgumentError("normal axis be a 3-vector"))
    end
    n_norm = sqrt(sum(n.^2))
    q = Quaternion(cos(θ / 2), sin(θ / 2) * n / n_norm)
    return rot3d_quat(v, q)
end

function rot3d_quat(v::Vector{T}, q::Quaternion) where T <: Real
    #println("using rot3D_quat->vector, q")
    if !(abs2(q) ≈ 1)
        throw(ArgumentError("quaternion must be a of unit length"))
    end
    qrot = q * Quaternion(0, v) * inv(q)
    vrot = [qrot.v1; qrot.v2; qrot.v3]
    return vrot
end
   
function rot3d_quat(V::AbstractMatrix{T}, q::Quaternion) where T <: Real
    #println("using rot3D_quat->matrix, q")
    if !(abs2(q) ≈ 1)
        throw(ArgumentError("quaternion must be a of unit length"))
    end
    Vr = similar(V)
    idx = 0
    for v in eachcol(V)
        idx = idx + 1
        qrot = q * Quaternion(0, collect(v)) * inv(q)
        Vr[1:end, idx] = [qrot.v1; qrot.v2; qrot.v3]
    end
    return Vr
end
```

```julia
pm(x) = x < 0 ? @sprintf(" - %.4f", abs(x)) : @sprintf(" + %.4f", abs(x))
```

```julia
function plot_quaternion_gui(θ, α, β, txt)
    n = [cos(α) * cos(β); sin(α) * cos(β); sin(β)]
    
    vr = rot3d_quat(v, n, θ)
    q = Quaternion(cos(θ / 2), sin(θ / 2) * n )

    plt = plot3d( vr[1, 1:end], vr[2, 1:end], vr[3, 1:end],
            xlim = (-rmax, rmax),
            ylim = (-rmax, rmax),
            zlim = (-rmax, rmax),
            aspect_ratio = :equal,
            label = "v", reuse=false,
            camera = (45, 50),
            xlabel="x", ylabel="y", zlabel="z",
    )
    
    # plot θ circle
    angles =  range(0, stop=2π, length=100)
    rt = cos(β)
    y = rt * cos.(angles)
    z = rt * sin.(angles)
    x = repeat([sin(β)], length(z))
    
    q_α = Quaternion(cos(α / 2), sin(α / 2) * [0, 0, 1] )
    q_β = Quaternion(cos(β / 2), sin(-β / 2) * [0, 1 ,0] )
    circle = rot3d_quat([x y z]', q_α * q_β)
    plot3d!(circle[1, 1:end], circle[2, 1:end], circle[3, 1:end], label="θ")
    
    # plot α segment
    angles =  range(0, stop=α, length=100)
    x = r * cos.(angles)
    y = r * sin.(angles)
    z = zeros(length(x))
    plot3d!([0; x; 0], [0; y;0 ], [0; z; 0], label="α")
    
    # plot β segment
    angles =  range(0, stop=β, length=100)
    x = r * cos.(angles)
    z = r * sin.(angles)
    y = zeros(length(x))
    seg = rot3d_quat([x y z]', q_α)
    plot3d!([0; seg[1, 1:end]; 0], [0; seg[2, 1:end]; 0], [0; seg[3, 1:end]; 0], label="β")
    
    txt[] = string("q = ", pm(q.s), pm(q.v1), "im", pm(q.v2), "jm", pm(q.v3), "km")
    
    # plot axis
    plot3d!([0, n[1]], [0, n[2]], [0, n[3]], label="n", color=:black)
    
end
```

```julia
angles =  -3.2:0.1:3.2
sliders = OrderedDict(:label => "quaternion demo", 
    :slider1=>slider(2*angles, value=0, label="θ"), 
    :slider2=>slider(angles, value=0, label="α"), 
    :slider3=>slider(angles, value=0, label="β"),
    :txt=>latex("")
    )

q = Quaternion(0)

output = Interact.@map( 
    plot_quaternion_gui(&sliders[:slider1], &sliders[:slider2], &sliders[:slider3], sliders[:txt])
    )
Interact.Widget(sliders, output = output)   
```
